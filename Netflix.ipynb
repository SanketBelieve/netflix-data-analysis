# Make a copy of the dataset

```python
df=netflix.copy()
```

```python
df.shape
```

Drop Null Values

```python
df = df.dropna()
df.shape
```

Print first 10 values

```python
df.head()
```

Convert Date Time format

```python
df["date_added"] = pd.to_datetime(df['date_added'])
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

# **Data Visualization**


```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,10))
sns.countplot(x='type', data=netflix)
plt.title('Distribution of Types')
plt.show()
```

Rating of shows and movie

```python
sns.set_style('whitegrid')
plt.figure(figsize=(13,13))

sns.countplot(x='rating', data=netflix)
plt.title('Distribution of Ratings')
plt.xticks(rotation=90, ha="right")

plt.show()
```

Relation Between Type And Rating

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=netflix)
plt.title('Relation between Type and Rating')
plt.show()
```

# Pie-chart for the Type: Movie and TV Shows

```python
labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

# **Pie Chart For Rating**

```python
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(25, 15))

# Generate the word cloud from a string with all country names
wordcloud = WordCloud(
    background_color='black',
    width=1920,
    height=1080
).generate(" ".join(df.country))

# Display the word cloud image on the axis
ax.imshow(wordcloud)

# Turn off the axis labels and ticks
ax.axis('off')

# Save the figure to a PNG file
fig.savefig('country.png')

# Show the final plot
plt.show()





```
